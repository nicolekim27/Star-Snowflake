```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
star_data = pd.read_excel(r'star_data.xlsx')
star_data = star_data[star_data.Prob > 90]
```

```python
mass = star_data['Mass']
age = star_data['Age']
plt.scatter(mass, age, color='r')
plt.xlabel('Solar Mass')
plt.ylabel('Age')
plt.show()
```

Looking at this plot, I see there are outliers past 8 solar masses and 0.7e10 year, so I will delete the values that have a mass greater than 8 solar masses and those that are older than 0.7e10 years. 

```python
star_data = star_data[star_data.Mass < 8]
star_data = star_data[star_data.Age < 7000000000]
```

```python
mass = star_data['Mass']
age = star_data['Age']
plt.scatter(mass, age, color='r')
plt.xlabel('Solar Mass')
plt.ylabel('Age')
plt.show()
```

My index values are out of order, so I am resetting it. Resetting the index created a new column with the old index values, so I will delete it

```python
star_data = star_data.reset_index()
del star_data['index']
```

```python
#I want to select 100 random entries
star_data_random = star_data.sample(n=100)
star_data_random
```

I replace star_data with star_data_random, reset the indices, and drop the newly created 'index' column. I re-create the age vs. mass plot to see if randomizing the values creates a different looking plot.

```python
star_data = star_data_random.reset_index()
del star_data['index']
mass = star_data['Mass']
age = star_data['Age']
plt.scatter(mass, age, color='r')
plt.xlabel('Solar Mass')
plt.ylabel('Age')
plt.show()
```

From this data, I can infer that there are not many stars more massive than our sun that are older than 1.0e9 years; or, younger stars are typically more massive than our sun. Or rather, the younger the star, the more massive it is / the older the star, the less massive it is (in comparison to our sun)<p>
     From these 100 values, I will create different plots. </p>

```python
#luminosity vs mass
luminosity = star_data['logL']
plt.scatter(mass, luminosity, color='r')
plt.xlabel('Solar Mass')
plt.ylabel('Solar Luminosity')
plt.show()
```

From this plot, we can infer, the more massive a star is, the more luminous it is, and vice versa.
