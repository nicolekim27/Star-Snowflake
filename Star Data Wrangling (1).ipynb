```python
import pandas as pd
import numpy as np
star_data = pd.read_excel(r'star_data.xlsx')
star_data
```

```python
#The first thing I noticed were negative values in the Age column. To fix this, I applied .abs() and was returned, “bad operand type for abs(): 'str'”. To convert those string values into integers, I used pd.to_numeric, and then I was able to apply .abs(). 
star_data.Age = pd.to_numeric(star_data.Age, errors='coerce')
star_data['Age'] = star_data['Age'].abs()
star_data
```

```python
#Then I deleted all rows containing a null, using dropna(). 
star_data = star_data.dropna()
star_data
```

```python
#The values in the column “LC” range between 1 and 35. LC stands for Luminosity Class. In astronomy, luminosity classes are labeled with Roman numerals from I to V. I deleted the column entirely because it does not accurately classify the objects into their correct luminosity class.
del star_data['LC']
star_data
```

```python
#I used list(set()) to delete any duplicates, but it appears there are none. 
list(set(star_data))
star_data
```

```python
#The Prob% column shows the probability of age and mass (goodness of fit). I want values with a high goodness of fit, so I removed rows with a Prob% of less than 90. 
#The ‘%’ was giving me trouble in my code so I changed the column title from ‘Prob%’ to ‘Prob’. 
star_data = star_data.rename(columns={"Prob%":"Prob"})
star_data[star_data.Prob > 90]
```
